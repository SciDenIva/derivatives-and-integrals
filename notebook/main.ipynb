```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-1, 5, 100)

plt.plot(x, x**2, label='x^2', color='g')
plt.plot(x, 4*x - 4, label='4x-4', color='r')

plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0, color='k')
plt.axhline(0, color='k')

plt.axvline(2, color='y', linestyle='--')
plt.axhline(4, color='y', linestyle='--')

plt.rc('axes.spines', **{'bottom':False, 'left':False, 'right':False, 'top':False})

plt.ylim((-1,10))
plt.xlim((-1,10))

plt.legend()

plt.savefig('pict_1.png', dpi=300)
```

```python
x = np.linspace(0, 20, 100)

plt.plot(x, 20 * np.sqrt(x), label='20*sqrt(x)', color='g')
plt.plot(x, 5*x + 20, label='5x+20', color='r')

plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0, color='k')
plt.axhline(0, color='k')

plt.axvline(4, color='y', linestyle='--')
plt.axhline(40, color='y', linestyle='--')

plt.rc('axes.spines', **{'bottom':False, 'left':False, 'right':False, 'top':False})

plt.ylim((0,60))
plt.xlim((0,20))

plt.legend()

plt.savefig('pict_3.png', dpi=300)
```

```python
x = np.linspace(0, 5, 100)

plt.plot(x, - x * x * x + 3 * x * x, label='-x^3 + 3x^2', color='g')

plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0, color='k')
plt.axhline(0, color='k')

plt.axvline(1.5, color='y', linestyle='--')
plt.axvline(2, color='r', linestyle='--')
plt.axvline(2.5, color='y', linestyle='--')

plt.annotate('P', (2.1,4.1))
plt.annotate('M', (1.3,3.5))
plt.annotate('N', (2.6,3.3))

plt.annotate('m', (1.52,0.1))
plt.annotate('n', (2.52,0.1))

plt.scatter(2,4,c='g')
plt.scatter(1.5,3.375,c='g')
plt.scatter(2.5,3.125,c='g')

plt.rc('axes.spines', **{'bottom':False, 'left':False, 'right':False, 'top':False})

plt.ylim((0,5))
plt.xlim((0,4))

plt.legend()

plt.savefig('pict_4.png', dpi=300)
```

```python
def func(x):
    return 1.3*(x-2)**3+2*(x-2)**2+2

def line(x, b):
    return x * (func(2.8) - func(1.5)) / (2.8 - 1.5) + b

x = np.linspace(0, 5, 100)

plt.plot(x, func(x),color='g')

plt.plot(x, line(x, 0.5),color='r', linestyle='--')
plt.plot(x, line(x, -0.67),color='r', linestyle='--')

plt.xlabel('x')
plt.ylabel('y')

plt.axvline(0, color='k')
plt.axhline(0, color='k')

plt.axvline(1.5, color='y', linestyle='--')
plt.axvline(2.8, color='y', linestyle='--')
plt.axvline(2.3, color='r', linestyle='--')

plt.axhline(func(1.5), color='y', linestyle='--')
plt.axhline(func(2.8), color='y', linestyle='--')

plt.annotate('A', (1.5 + 0.1, func(1.5) + 0.3))
plt.annotate('B', (2.8 + 0.1, func(2.8) - 0.2))
plt.annotate('C', (2.3 + 0.1, func(2.3) - 0.2))

#plt.annotate('m', (1.52,0.1))
#plt.annotate('n', (2.52,0.1))

plt.scatter(1.5,func(1.5),c='g')
plt.scatter(2.8,func(2.8),c='g')
c = 2.3
plt.scatter(c,func(c),c='r')

#plt.scatter(2.5,3.125,c='g')

plt.rc('axes.spines', **{'bottom':False, 'left':False, 'right':False, 'top':False})

plt.ylim((0,5))
plt.xlim((0,4))

#plt.legend()

plt.savefig('pict_5.png', dpi=300)
```
